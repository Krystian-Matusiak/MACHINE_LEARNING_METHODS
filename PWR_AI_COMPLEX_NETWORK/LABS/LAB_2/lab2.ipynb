```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import collections
```

## Erdos Renyi Graph

```python
N = 100
p = 0.05

ER_G = nx.erdos_renyi_graph(N, p, directed=False)
```

```python
nx.draw(ER_G, with_labels=True)
```

```python
nx.draw_circular(ER_G)
```

```python
degrees = sorted([d for _, d in ER_G.degree()], reverse=True)  # degree sequence
degreeCount = collections.Counter(degrees)
deg, cnt = zip(*degreeCount.items())


plt.bar(deg, cnt, width=0.80, color='b')
plt.xlabel("Degree")
plt.ylabel("Number of samples")
```

```python
# con_components = [each for each in nx.connected_components(ER_G)]
# con_components = con_components[0]
# con_components
```

```python
d = dict(ER_G.degree)
keys = [each for each, _ in d.items()]
PR = [(1200*pr)**2 for _,pr in nx.pagerank(ER_G).items()]
nx.draw(ER_G, nodelist=keys, node_size=PR)
```

```python
nx.diameter(ER_G)
```
